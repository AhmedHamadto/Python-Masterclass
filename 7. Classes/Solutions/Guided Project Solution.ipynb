```python
# Step 1: Create the Product Class
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def calculate_total_price(self):
        return self.price * self.quantity

# Step 2: Create Product Categories
class Electronics(Product):
    def __init__(self, name, price, quantity, brand):
        super().__init__(name, price, quantity)
        self.brand = brand

    def display_info(self):
        print(f"Name: {self.name}, Price: ${self.price:.2f}, Brand: {self.brand}")

class Clothing(Product):
    def __init__(self, name, price, quantity, size):
        super().__init__(name, price, quantity)
        self.size = size

    def display_info(self):
        print(f"Name: {self.name}, Price: ${self.price:.2f}, Size: {self.size}")

class Food(Product):
    def __init__(self, name, price, quantity, expiry_date):
        super().__init__(name, price, quantity)
        self.expiry_date = expiry_date

    def display_info(self):
        print(f"Name: {self.name}, Price: ${self.price:.2f}, Expiry Date: {self.expiry_date}")

# Step 3: Create the Shopping Cart Class
class ShoppingCart:
    def __init__(self):
        self.items = []
        self.total = 0

    def add_item(self, product):
        self.items.append(product)
        self.total += product.calculate_total_price()

    def display_cart(self):
        for item in self.items:
            item.display_info()
        print(f"Total Cost: ${self.total:.2f}")

# Step 6: Interaction
electronics = Electronics("Laptop", 1000, 2, "Dell")
clothing = Clothing("T-Shirt", 20, 5, "M")
food = Food("Bread", 3, 10, "2023-12-31")

cart = ShoppingCart()
cart.add_item(electronics)
cart.add_item(clothing)
cart.add_item(food)

print("Shopping Cart Contents:")
cart.display_cart()
```
